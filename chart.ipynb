```python
import pickle

with open("nn-all-loss.pkl", 'rb') as fp:
    nn_all_loss = pickle.load(fp)
with open("nn-cont-loss.pkl", 'rb') as fp:
    nn_cont_loss = pickle.load(fp)
```

```python
import matplotlib.pyplot as plt

x_vales = range(0, len(nn_all_loss))
plt.plot(x_vales, nn_all_loss, label='All variables')
plt.plot(x_vales, nn_cont_loss, label='Cont variables')
plt.legend()
plt.title("Neural Network Training Losses")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.show()
```

```python
# acc vs time plot

# TODO time not all measures
labeled = {
    "Continous bayes": (1929, 84.02),
    "All bayes": (2657, 79.56),
    "Continous logistic regression": (3.2, 92.39),
    "All logistic regression": (3.7, 93.06),
    "Continous Neural Net": (185.1, 98.96),
    "All Neural Net": (189.7, 98.96)

}

# Extracting x and y coordinates
x_values = [value[0] for value in labeled.values()]
y_values = [value[1] for value in labeled.values()]
labels = labeled.keys()

# Plot
plt.figure(figsize=(6, 6))
plt.scatter(x_values, y_values, marker='o', color='blue')

# Adding labels
sign = 1
for label, x, y in zip(labels, x_values, y_values):
    sign *= -1
    plt.annotate(label, (x, y), textcoords="offset points", xytext=(sign * 10,sign * 10), ha='center')

# Adding title and labels
plt.title('Prediction Time vs Accuracy')
plt.xlabel('Prediction Time (in ms)')
plt.ylabel('Balanced Accuracy %')

# Display the plot
plt.grid(True)

# plt.gca().invert_xaxis()
plt.show()

```
